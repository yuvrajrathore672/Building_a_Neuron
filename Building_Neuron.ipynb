```python
import torch 
import torch.nn as nn
```

```python
# input 
torch.manual_seed(42)
input = torch.tensor([1.0,2.0,3.0])
```

```python
neuron = nn.Linear(in_features=3,out_features=1)
output = neuron(input)
output
```

```python
# verify 
print(neuron.weight)
print(neuron.bias)

neuron.weight @ input + neuron.bias
```
